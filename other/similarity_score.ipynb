```python
import os
import pickle
from openbabel import openbabel as ob
from openbabel import pybel as pb
from rdkit.Chem import AllChem, Descriptors, Draw
from rdkit import Chem
from rdkit import DataStructs
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def generate_morgan_fp(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 3, nBits=2048)  # Morgan fingerprint radius=2, 1024-bit vector
    return fp

def calculate_tanimoto_similarity(fp1, fp2):
    if fp1 is None or fp2 is None:
        return None
    similarity = DataStructs.TanimotoSimilarity(fp1, fp2)
    return similarity
```

```python
# reading fingerprints

# pdb
f = open("anshul/fingerprints/morgan/pdb.txt", 'rb')
pdb_fps = pickle.load(f)
f.close()

# drugbank
f = open("anshul/fingerprints/morgan/drugbank.txt", 'rb')
db_fps = pickle.load(f)
f.close()
```

# denovo

```python
prot_name = "2g94"
lig_dir = f"anshul/denovo/results/{prot_name}"
ligs = [name for name in os.listdir(lig_dir) if name.endswith("pdbqt")]

ts = []
for name in ligs:
    fp = generate_morgan_fp(next(pb.readfile("pdbqt", os.path.join(lig_dir, name))).write('smi'))
    if fp is None : continue
    ts.append([])
    for e in pdb_fps:
        ts[-1].append(calculate_tanimoto_similarity(fp, e[1]))
    
    for e in db_fps:
        ts[-1].append(calculate_tanimoto_similarity(fp, e[1]))

```

```python
ts = np.array(ts)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.kdeplot(ts.mean(axis=1), ax=axes[0])
axes[0].set_title(f'Mean {prot_name}')

sns.kdeplot(ts.max(axis=1), ax=axes[1], color="red")
axes[1].set_title(f'Max {prot_name}')


plt.tight_layout() 
plt.show()
```
