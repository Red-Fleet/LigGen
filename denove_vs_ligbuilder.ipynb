```python
import matplotlib.pyplot as plt
import random
import math
import re
import os
import numpy as np
import seaborn as sns
```

2g94

```python
name = "2g94"

f = open(f"results_ligbuilder/{name}/scores.txt")
energy_bu = [float(s.split()[1]) for s in f.readlines() if len(s.split())>0 and len(s)>0]
f.close()

f = open(f"results/ligbuilder/{name}_scores.txt")
energy_de = [float(l.split(':')[1]) for l in f.readlines() if l.split(':')[1] != 'None']
f.close()


# Create histogram
sns.histplot(energy_bu, bins=20, kde=False, color="red", edgecolor='black', alpha=1, label='ligbuilder 3')  
sns.histplot(energy_de, bins=30, kde=False, color="blue", edgecolor='black', alpha=0.5, label='de novo monte carlo')
# Add labels and title
# Add labels and title
plt.xlabel('ENERGY')
plt.ylabel('COUNT')
plt.title(f'2G94: Beta-Secretase')
plt.legend()
# Show plot
plt.show()
```
